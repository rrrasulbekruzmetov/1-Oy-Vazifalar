# Mode() YORDAMIDA "GENDER" USTUNINI TO`LDIRAMIZ 

```python
df2['Gender'].fillna(df2['Gender'].mode()[0], inplace=True)
df2
```

```python
df2.isnull().sum()
```

# Median() YORDAMIDA "AGE " USTUNINI TO`LDIRAMIZ 

```python
df2['Age'].fillna(df2['Age'].median(), inplace=True)
df2
```

```python
df2.isnull().sum()
```
