# Data Setda tushirib qoldirilgan ustunlarni aniqlaymiz 

```python
df.isnull().sum()
```

# Mean (numerical) - orqali to`ldirib ko`ramiz  // DEPTH ERROR USTUNI 

```python
df['Depth Error'].fillna(df['Depth Error'].mean(), inplace=True)
```

# Natijani ko`ramiz

```python
df.isnull().sum()
```

# Mode - orqali to`ldirib ko`ramiz // Azimuthal Gap USTUNI 

```python
df['Azimuthal Gap'].fillna(df['Azimuthal Gap'].mode()[0], inplace=True)
```

# Natijani ko`ramiz

```python
df.isnull().sum()
```

# Median - orqali to`ldirib ko`ramiz // Horizontal Distance USTUNI 

```python
df['Horizontal Distance'].fillna(df['Horizontal Distance'].median(), inplace=True)
```

# Natijani ko`ramiz

```python
df.isnull().sum()
```
